**Importing the libraries and reading the csv**

```python
import numpy as np
import pandas as pd
```

```python
videos = pd.read_csv("/content/INvideos.csv")
```

```python
videos.head(10)
```

```python
videos.info()
```

**Data Cleaning**

```python
#missing and nan values
videos.isnull().sum()
```

```python
#duplicates
videos.duplicated().sum()
```

```python
#dropping only duplicates
videos.drop_duplicates(inplace=True)
```

```python
#setting trending date to datetime
videos['trending_date'] = pd.to_datetime(videos['trending_date'], format='%y.%d.%m')
```

```python
#setting publish time to datetime
videos['publish_time'] = pd.to_datetime(videos['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
videos.dtypes
```

```python
columns_to_drop = ['video_id', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed', 'description']
videos = videos.drop(columns=columns_to_drop)
display(videos.head())
```

```python
#Distribution of views, likes, dislikes, and comments
numeric_columns = ['views', 'likes', 'dislikes', 'comment_count']
videos[numeric_columns].describe()
```

```python
# Top trending channels analysis
top_channels = videos['channel_title'].value_counts().head(10)
print(top_channels)
```

```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plotting top trending channels
#using horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_channels.values, y=top_channels.index, palette='viridis')
plt.xlabel('Number of Videos')
plt.ylabel('Channel Title')
plt.title("Top Trending channels")
plt.show()
```

```python
#Most common video categories.
common_categories = videos['category_id'].value_counts().head(10)
print(common_categories)
```

```python
#plotting a bar chart
plt.figure(figsize = (10,6))
sns.barplot(x = common_categories.values, y = common_categories.index, color = 'seagreen')
plt.xlabel('Number of videos')
plt.ylabel('Category')
plt.title('Most common video categories')
plt.show()
```

```python
#Time trends of publish and trending dates.
# so extraacting the time components - day, month, year
# trending date , publish are already in datetime format
```

```python
videos['trending_date'] = pd.to_datetime(videos['trending_date'], format='%y.%d.%m')
videos['trending_year'] = videos['trending_date'].dt.year
videos['trending_month'] = videos['trending_date'].dt.month
videos['trending_day_of_week'] = videos['trending_date'].dt.dayofweek

videos['publish_year'] = videos['publish_time'].dt.year
videos['publish_month'] = videos['publish_time'].dt.month
videos['publish_day_of_week'] = videos['publish_time'].dt.dayofweek
```

```python
# trending dates trends
trending_monthly_counts = videos.groupby(['trending_year', 'trending_month']).size().reset_index(name='video_count')
trending_monthly_counts['year_month'] = trending_monthly_counts['trending_year'].astype(str) + '-' + trending_monthly_counts['trending_month'].astype(str)
display(trending_monthly_counts.head())
```

```python
#line plot to visualize the monthly trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='video_count', data=trending_monthly_counts)
plt.xlabel('Year-Month')
plt.ylabel('Number of Trending Videos')
plt.title('Monthly Trend of Trending Videos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#weekly trending videos
trending_dayofweek_counts = videos.groupby('trending_day_of_week').size().reset_index(name='video_count')
display(trending_dayofweek_counts.head())
```

```python
# visualize the day of the week trending videos
plt.figure(figsize=(10, 6))
sns.barplot(x='trending_day_of_week', y='video_count', data=trending_dayofweek_counts, color='skyblue')
plt.xlabel('Day of the Week (0=Monday, 6=Sunday)')
plt.ylabel('Number of Trending Videos')
plt.title('Trending Videos by Day of the Week')
plt.show()
```

```python
#count of the videos by month
publish_monthly_counts = videos.groupby(['publish_year', 'publish_month']).size().reset_index(name='video_count')
publish_monthly_counts['year_month'] = publish_monthly_counts['publish_year'].astype(str) + '-' + publish_monthly_counts['publish_month'].astype(str)
display(publish_monthly_counts.head())
```

```python
#line plot to visualize the monthly trend of published videos
plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='video_count', data=publish_monthly_counts)
plt.xlabel('Year-Month')
plt.ylabel('Number of Published Videos')
plt.title('Monthly Trend of Published Videos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#count of the published videos in day of week
publish_dayofweek_counts = videos.groupby('publish_day_of_week').size().reset_index(name='video_count')
display(publish_dayofweek_counts.head())
```

```python
#visualize using the bar plot for published videos
plt.figure(figsize=(10, 6))
sns.barplot(x='publish_day_of_week', y='video_count', data=publish_dayofweek_counts, color='coral')
plt.xlabel('Day of the Week (0=Monday, 6=Sunday)')
plt.ylabel('Number of Published Videos')
plt.title('Published Videos by Day of the Week')
plt.show()
```

```python
# Engagement metrics by category/channel
```

```python
#How does engagement (likes/comments) vary by category?
# Calculate average engagement metrics by category
category_engagement = videos.groupby('category_id')[['views', 'likes', 'dislikes', 'comment_count']].mean().reset_index()
display(category_engagement.head())
```

```python
# Calculate average engagement metrics by channel
channel_engagement = videos.groupby('channel_title')[['views', 'likes', 'dislikes', 'comment_count']].mean().reset_index()
# Display top channels by average views
top_channels_by_views = channel_engagement.sort_values(by='views', ascending=False).head(10)
display(top_channels_by_views)
```

```python
# Visualize average views by category
plt.figure(figsize=(12, 6))
sns.barplot(x='views', y='category_id', data=category_engagement.sort_values(by='views', ascending=False).head(10), orient='h', palette='coolwarm')
plt.xlabel('Average Views')
plt.ylabel('Category ID')
plt.title('Average Views by Category')
plt.show()
```

```python
# Visualize average likes by category
plt.figure(figsize=(12, 6))
sns.barplot(x='likes', y='category_id', data=category_engagement.sort_values(by='likes', ascending=False).head(10), orient='h', palette='viridis')
plt.xlabel('Average Likes')
plt.ylabel('Category ID')
plt.title('Average Likes by Category')
plt.show()
```

```python
# Visualize average views for top channels
plt.figure(figsize=(12, 6))
sns.barplot(x='views', y='channel_title', data=top_channels_by_views, orient='h', palette='plasma')
plt.xlabel('Average Views')
plt.ylabel('Channel Title')
plt.title('Average Views for Top 10 Channels')
plt.show()
```

```python
# What is the correlation between views, likes, comments, and dislikes?
# Calculate correlation matrix
correlation_matrix = videos[['views', 'likes', 'dislikes', 'comment_count']].corr()
display(correlation_matrix)
```

```python
# Calculate the time difference between trending date and publish time
videos['time_to_trend'] = (videos['trending_date'] - videos['publish_time'].dt.tz_localize(None)).dt.days
display(videos[['title', 'publish_time', 'trending_date', 'time_to_trend']].head())
```

```python
#plotting correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='inferno', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## Summary of Findings:

Based on the exploratory data analysis of the YouTube trending videos dataset for India, here are the key findings:

### Data Overview and Cleaning:
- The dataset contains information about trending YouTube videos in India, including video details, trending dates, publish times, and engagement metrics.
- Missing values were identified primarily in the 'description' column (561 missing values).
- Duplicate rows were found and removed from the dataset (4263 duplicates).
- The 'trending_date' and 'publish_time' columns were successfully converted to datetime objects for time-based analysis.
- Several columns (`video_id`, `thumbnail_link`, `comments_disabled`, `ratings_disabled`, `video_error_or_removed`, `description`) were dropped as they were deemed less relevant for the current analysis goals.

### Distribution of Engagement Metrics:
- The distribution of views, likes, dislikes, and comment counts shows a wide range, with some videos achieving very high engagement while many others have significantly lower numbers. The descriptive statistics highlight the skewness in these metrics.

### Top Trending Channels and Categories:
- The analysis identified the top 10 channels with the most trending videos, indicating the prominent content creators in the Indian YouTube space during the dataset's timeframe.
- The most common video categories were also identified, showing which types of content are most frequently appearing on the trending list. Category 24 (Entertainment) and 25 (News & Politics) were the most frequent.

### Time Trends of Publish and Trending Dates:
- The monthly trend of trending videos showed some variation but no clear overall upward or downward trend between late 2017 and mid-2018.
- The number of videos trending was relatively consistent across the days of the week, suggesting no strong weekly pattern for a video to appear on the trending list.
- Similar analyses were performed for published videos, showing their distribution over months and days of the week.

### Engagement Metrics by Category and Channel:
- Analysis of average engagement metrics (views, likes, dislikes, comment counts) by category revealed that certain categories tend to have higher average engagement than others.
- Examining average engagement metrics for individual channels highlighted that some channels consistently achieve very high views and likes, indicating their popularity and influence.

### Time to Trend:
- The calculation of `time_to_trend` (difference between trending date and publish time) provides insight into how quickly videos, on average, reach the trending page after being published. Further analysis of this metric could reveal factors influencing the speed of trending.

### Overall Insights:
- The analysis provides a foundational understanding of the characteristics of trending videos on YouTube India during the dataset's period.
- While no strong seasonal or weekly patterns were observed for trending dates themselves, the analysis of categories and channels points to the importance of content type and creator popularity in achieving trending status.
- The correlation analysis showed moderate to strong positive correlations between views, likes, and comment counts, which is expected as these metrics often increase together for popular videos.

### Next Steps:
- Further investigation into the factors influencing `time_to_trend` could be valuable.
- Analyzing the content of the videos (e.g., title keywords, tags) could provide deeper insights into what makes a video trend.
- Exploring the relationship between publish time of day and trending could reveal more granular temporal patterns.
- Building predictive models to forecast which videos are likely to trend could be a potential application of this analysis.
